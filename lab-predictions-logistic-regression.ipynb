```python
# Create a query or queries to extract the information you think may be relevant for building the prediction model. 
#It should include some film features and some rental features (X).

connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)

query = '''SELECT DISTINCT f.title,
MAX(IF(r.rental_date LIKE '%%2005-05%%', 1, 0)) AS rented_in_may
FROM film f
LEFT JOIN inventory i ON f.film_id = i.film_id
LEFT JOIN rental r ON i.inventory_id = r.inventory_id
GROUP BY f.title
;'''

data = pd.read_sql_query(query, engine)
data.head(10)
```

```python
#1.Create a query to get the list of all unique film titles and a boolean indicating if it was rented (rental_date) in May 2005. 
#2.Create new column called - 'rented_in_may'). This will be our TARGET (y) variable.
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)

query1='''SELECT f.title, f.rating, f.rental_rate, f.length, f.rental_duration
FROM film f
ORDER BY title;'''

features = pd.read_sql_query(query1, engine)
features
```

```python
#3.Read the data into a Pandas dataframe. At this point you should have 1000 rows. 
#Number of columns depends on the number of features you chose.
final_data = pd.concat([features,data['rented_in_may']], axis=1)
final_data
```

```python
#4.Analyze extracted features (X) and transform them. You may need to encode some categorical variables, or scale numerical variables.
```

```python
# X/y split
X = final_data.drop(['title','rented_in_may'], axis = 1)
y = final_data['rented_in_may']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
print('X_train shape is:',X_train.shape)
print('y_train shape is:', y_train.shape)
print('X_test shape is:', X_test.shape)
print('y_train shape is:', y_test.shape)
```

```python
X_train_num = X_train.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)
X_train_num
X_train_cat
```

```python
X_test_num = X_test.select_dtypes(np.number)
X_test_cat = X_test.select_dtypes(object)
X_test_num
X_test_cat
```

```python
from sklearn.preprocessing import MinMaxScaler
# Scaling numerical data
transformer = MinMaxScaler().fit(X_train_num) 
X_train_normalized = transformer.transform(X_train_num)
X_train_normalized = pd.DataFrame(X_train_normalized, columns=X_train_num.columns)
X_train_normalized.head()
```

```python
X_test_normalized = transformer.transform(X_test_num)
X_test_normalized = pd.DataFrame(X_test_normalized, columns=X_test_num.columns)
X_test_normalized.head()
```

```python
#Encoding the categoricals
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder().fit(X_train_cat)
print(encoder.categories_)
print(encoder.categories_[0])
encoded = encoder.transform(X_train_cat).toarray()
X_train_encoded = pd.DataFrame(encoded,columns=encoder.categories_[0])
X_train_encoded.head()
```

```python
encoded = encoder.transform(X_test_cat).toarray()
X_test_encoded = pd.DataFrame(encoded,columns=encoder.categories_[0])
X_test_encoded.head()
X_test_encoded.isna().sum()
```

```python
X_train_transformed = pd.concat([X_train_normalized, X_train_encoded], axis=1)
X_train_transformed.head()
```

```python
X_test_normalized.reset_index(drop=True, inplace=True)
X_test_encoded.reset_index(drop=True, inplace=True)
X_test_transformed = pd.concat([X_test_normalized, X_test_encoded],axis=1)
X_test_transformed.head()
```

```python
#5.Create a logistic regression model to predict 'rented_in_may' from the cleaned data.
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression().fit(X_train_transformed, y_train)
print('training set score:{:3f}'.format(logreg.score(X_train_transformed,y_train)))
print('test set score:{:3f}'.format(logreg.score(X_test_transformed,y_test)))
```

```python
from sklearn.metrics import confusion_matrix
prediction = logreg.predict(X_test_transformed)
confusion_matrix(y_test,prediction) 
#good news is: we have no False negatives, bad news is the model only predicted 1.
```
